```python
import requests
from bs4 import BeautifulSoup
```

```python
# 2. html 요소 얻기
# 유의할 점 - m.으로 시작하는 모바일 주소로 변환
blog1_url = "https://m.blog.naver.com/a970830/223348867173"
page1= requests.get(blog1_url)

blog2_url = "https://m.blog.naver.com/peony_fuzzy/223359551371"
page2= requests.get(blog2_url)

blog3_url = "https://m.blog.naver.com/ckalswn12/223367644916"
page3= requests.get(blog3_url)
```

```python
page3
```

```python
# 3. BeatifulSoup 으로 파싱
bs_result1 = BeautifulSoup(page1.text, 'html.parser')
bs_result1

bs_result2 = BeautifulSoup(page2.text, 'html.parser')
bs_result2

bs_result3 = BeautifulSoup(page3.text, 'html.parser')
bs_result3
```

```python
bs_result3
```

```python
# 4. 요소 추출
result1 = bs_result1.find_all("span", class_="se-fs-")
result1_list = []

for item in result1:
    result1_list.append(item.text)
    #print(item.text)


print(result1_list)
```

```python
# 4. 요소 추출
result2 = bs_result2.find_all("span", class_="se-fs-")
result2_list = []

for item in result2:
    result2_list.append(item.text)
    #print(item.text)


print(result2_list)
```

```python
# 4. 요소 추출
result3 = bs_result3.find_all("span", class_="se-fs-")
result3_list = []

for item in result3:
    result3_list.append(item.text)
    #print(item.text)


print(result3_list)
```

```python
last_result = [
                {"블로그이름" : "blog1", "블로그 내용" : result1_list},
                {"블로그이름" : "blog2", "블로그 내용" : result2_list},
                {"블로그이름" : "blog3", "블로그 내용" : result3_list}
            ]
# 데이터 프레임
# 시각화
```

```python
import pandas as pd
data = pd.DataFrame(last_result)
data
data.to_csv
```
