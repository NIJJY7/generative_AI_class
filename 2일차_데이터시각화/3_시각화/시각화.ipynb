```python
plt.title("선그래프")
plt.plot([10,20,30,40,50], [2,4,4,3,10])
plt.show()
```

```python
plt.title("선그래프")
plt.plot([10,20,30,40,50], [2,4,4,3,10], c='r', lw =4, ls = '--', marker = "o", ms = 8, mec = 'g', mew =3)
plt.xlabel("일", labelpad= 20)
plt.ylabel("가격", rotation=0, labelpad=20)
plt.show()
```

```python
plt.title("선그래프", size = 20)
y = [2,4,6,8,10]
x = [10, 20, 30, 40, 50]
x_names = ['1일', '2일', '3일', '4일', '5일']
plt.plot(x,y)

plt.xticks(x,x_names, rotation=45)
plt.grid(True)
plt.xlabel("일")
plt.ylabel("가격", rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (5,6))
plt.title("매출 비교 그래프", size = 20)
plt.plot([2,7,5,2], label = '게임', c = 'r')
plt.plot([2,5,9,2], label = '게임2', c = 'b')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (5,6))

plt.subplot(2,4,1)  #1행 2열의 첫번째 그래프
plt.title("매출 비교 그래프", size = 20)
plt.plot([2,7,5,2], label = '게임', c = 'r')

plt.subplot(2,4,2)  #1행 2열의 두번째 그래프
plt.plot([2,5,9,2], label = '게임2', c = 'b')

plt.subplot(2,4,8)  #1행 2열의 두번째 그래프
plt.plot([2,5,9,2], label = '게임2', c = 'b')
plt.legend()
plt.show()
```
